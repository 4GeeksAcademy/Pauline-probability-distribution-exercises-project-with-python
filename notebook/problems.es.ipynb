# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

DISTRIBUCIÓN NORMAL

```python
# TODO

from scipy.stats import norm

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

#Calculamos de probabilidad en una distribución normal 
# Tenemos que usar .pdf porque pmf que ha sido usada x la binomial, es solo para distrib de variables aleatorias, discretas :

def dnorm(x, media=0, std=1):
    """
    Calculamos la densidad de probabilidad de la distribución normal.
    """
    result = norm.pdf(x, loc=media, scale=std)
    return result

#Haceos una prueba con x = 1 :

x = 1
resultado = dnorm(x)

print(f"La densidad de probabilidad en x = {x} es: {resultado}")

```

```python
def pnorm(p, media=0, std=1):
    """
    Calcula el acumulado de la distribución normal.
    """
    result = norm.cdf(p, loc=media, scale=std)
    return result


#Hacemos una prueba con p = 1 :

p = 1
resultado = pnorm(p)

print(f"La probabilidad acumulada hasta un valor p = {p} es: {resultado}")
```

```python
def qnorm(q, media=0, std=1):
    """
    Calcula la función cuantil a partir de la distribución normal.
    """
    result = norm.ppf(q, loc=media, scale=std)

    return result


#Haceos una prueba con q = 0.25 (significa que estamos buscando un valor x tal que el 25% de los valores en la distribución estén por debajo de x) :

q = 0.25 
resultado = qnorm(q)
print(f"El valor de x tal que la probabilidad acumulada hasta x es {q} es: {resultado}")
```

```python
def rnorm(n, media=0, std=1):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = norm.rvs(loc=media, scale=std, size=n)

    return result


# Ejemplo con n=5:
n = 5
resultados = rnorm(n)

print(f"Debajo imprimimos {n} números aleatorios generados con una distribución normal estándar:")
print(resultados)
```

```python
#Distribución normal(10, 0.2), donde 10 es la media, y 0,2 la std :


media = 10
std = 0.2

np.random.seed(42)


x=10
resultado = dnorm(x, media, std) 
print(f"Probabilidad que la función normal (10,0.2) tenga el valor {x} es: {resultado}")

p = 10
resultado = pnorm(p, media, std)
print(f"Probabilidad que una distribución normal (10, 0.2) tenga un valor menos o igual a p = {p} es: {resultado}")


q = 0.9
resultado = qnorm(q, media, std)
print(f"El valor de x tal que la probabilidad acumulada hasta x es {q} es: {resultado}")


n = 2000
resultados = rnorm(n, media, std)
print(f"Genera {n} números aleatorios generados con una distribución normal (10, 0.2) : {resultados}:")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

media = 0
std = 1
data = np.arange(-5, 5, 0.01)

pdf = dnorm(data, media, std)
cdf = pnorm(data, media, std)
ppf = qnorm(data, media, std) 



fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, pdf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# TODO

from scipy.stats import uniform

# Definir todas las funciones de probabilidad relacionadas con esta distribución.
# Representamos la fonción de densidad de probabilidad :

def duniform(x, a=0, b=10):
    """
    Calculamos la densidad de probabilidad de la distribución uniforme en el intervalo [a, b].
    """
    result = uniform.pdf(x, loc=a, scale=(b - a))  # loc = inicio del intervalo, y scale = longitud del intervalo
    return result

# Ejemplo con x = 1, intervalo [0, 10]:

x = 1
a = 0
b = 10
resultado = duniform(x, a, b)

print(f"La densidad de probabilidad en x = {x} para la distribución uniforme en el intervalo [{a}, {b}] es: {resultado}")


```

```python
def puniform(p, a=0, b=10):
    """
    Calcula el acumulado de la distribución normal.
    """
    result = uniform.cdf(p, loc=a, scale= b-a)
    return result


#Hacemos una prueba con p = 10 :

p = 4

resultado = puniform(p)

print(f"La probabilidad de que el resultado de una distribución uniforme tenga un valor menos o igual {p} en el intervalo [{a}, {b}] es: {resultado}")
```

```python
def quniform(q, a=0, b=10):
    """
    Calcula la función cuantil a partir de la distribución normal.
    """
    result = uniform.ppf(q, loc=a, scale= b-a)

    return result


#Haceos una prueba con q = 0.25 (significa que estamos buscando un valor x tal que el 25% de los valores en la distribución estén por debajo de x) :

q = 0.9 
resultado = quniform(q)
print(f"El valor del resultado tal que la probabilidad acumulada hasta ese resultado es {q} es: {resultado}")
```

```python
def runiform(n, a=0, b=10):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = uniform.rvs(loc=a, scale= b-a, size=n)

    return result


# Ejemplo con n=2000:
n = 2000
resultados = runiform(n)

print(f"Debajo imprimimos {n} números aleatorios generados con una distribución uniforme:")
print(resultados)
```
